```python
import requests
import pandas as pd


# API URL
api_url = "https://public.opendatasoft.com/api/explore/v2.1/catalog/datasets/medecins/records?limit=20"

# Send a GET request to the API
response = requests.get(api_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Retrieve the JSON data from the response
    data = response.json()
    
    # Convert the JSON data into a Pandas DataFrame
    df = pd.DataFrame(data)
    
    # Display the DataFrame
    print(df)
else:
    # Display an error message if the request failed
    print("Error:", response.status_code)
```

```python
results_df = pd.json_normalize(df['results'])

# Combine 'total_count' column with the newly created DataFrame
result_with_total_count = pd.concat([df['total_count'], results_df], axis=1)

# Display the combined DataFrame
print(result_with_total_count)
```
